## ðŸ’¾ The data: sleep_health_data.csv

SleepInc has provided you with an anonymized dataset of sleep and lifestyle metrics for 374 individuals. This dataset contains average values for each person calculated over the past six months. The data is saved as `sleep_health_data.csv`.

The dataset includes 13 columns covering sleep duration, quality, disorders, exercise, stress, diet, demographics, and other factors related to sleep health. 

| Column | Description |
|---------|----------------------------------------|  
| `Person ID` | An identifier for each individual. |
| `Gender` | The gender of the person (Male/Female). |  
| `Age` | The age of the person in years. |
| `Occupation` | The occupation or profession of the person. |
| `Sleep Duration (hours)` | The average number of hours the person sleeps per day. |
| `Quality of Sleep (scale: 1-10)` | A subjective rating of the quality of sleep, ranging from 1 to 10. |
| `Physical Activity Level (minutes/day)` | The average number of minutes the person engages in physical activity daily. |  
| `Stress Level (scale: 1-10)` | A subjective rating of the stress level experienced by the person, ranging from 1 to 10. |
| `BMI Category` | The BMI category of the person (e.g., Underweight, Normal, Overweight). |
| `Blood Pressure (systolic/diastolic)` | The average blood pressure measurement of the person, indicated as systolic pressure over diastolic pressure. |
| `Heart Rate (bpm)` | The average resting heart rate of the person in beats per minute. |
| `Daily Steps` | The average number of steps the person takes per day. |
| `Sleep Disorder` | The presence or absence of a sleep disorder in the person (None, Insomnia, Sleep Apnea). |

```python
# Start coding here
# Use as many cells as you need
import pandas as pd
```

```python
sleep_health_data = pd.read_csv('sleep_health_data.csv')
sleep_health_data
```

```python
lowest_sleep_occ = sleep_health_data.groupby('Occupation')['Sleep Duration'].mean()
lowest_sleep_occ = lowest_sleep_occ.sort_values().index[0]
lowest_sleep_occ
```

The first two solutions didnt work because the output was not a string variable. it was a panda series. With this code, grouping the occupations, finding the average for each occupation, and then sorting values in ascending order, then selecting the first index(because the occupations are the indexes due to groupby) ; 'index[0]', gives us a string variable 

```python
lowest_sleep_quality_occ = sleep_health_data.groupby('Occupation')['Quality of Sleep'].mean()
lowest_sleep_quality_occ = lowest_sleep_quality_occ.sort_values().index[0]
lowest_sleep_quality_occ
```

```python
Normal = sleep_health_data[sleep_health_data['BMI Category']== 'Normal']
total_normal = len(Normal)
total_normal
```

```python
Normal_Insomnia = sleep_health_data[(sleep_health_data['Sleep Disorder']== 'Insomnia') & (sleep_health_data['BMI Category'] == 'Normal')]
Normal_Insomnia
```

```python
Overweight = sleep_health_data[sleep_health_data['BMI Category']== 'Overweight']
total_overweight = len(Overweight)
total_overweight
```

```python
Overweight_Insomnia = sleep_health_data[(sleep_health_data['Sleep Disorder']== 'Insomnia') & (sleep_health_data['BMI Category'] == 'Overweight')]
Overweight_Insomnia
```

```python
Obese = sleep_health_data[sleep_health_data['BMI Category']== 'Obese']
total_obese = len(Obese)
total_obese
```

```python
Obese_Insomnia = sleep_health_data[(sleep_health_data['Sleep Disorder']== 'Insomnia') & (sleep_health_data['BMI Category'] == 'Obese')]
Obese_Insomnia
```

```python
Normal_Insomnia_Ratio = round(len(Normal_Insomnia)/total_normal, 2)
Normal_Insomnia_Ratio
```

```python
Overweight_Insomnia_Ratio = round(len(Overweight_Insomnia)/total_overweight, 2)
Overweight_Insomnia_Ratio
```

```python
Obese_Insomnia_Ratio = round(len(Obese_Insomnia)/total_obese, 2)
Obese_Insomnia_Ratio
```

```python
bmi_insomnia_ratios = {'Normal': Normal_Insomnia_Ratio,
                       'Overweight' : Overweight_Insomnia_Ratio,
                        'Obese': Obese_Insomnia_Ratio}
bmi_insomnia_ratios
```
